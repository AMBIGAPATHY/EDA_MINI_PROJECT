# Task 9: Which Country has the highest no. of Tv shows

# i)Show all the countries and no. of tv shows in a graph as well

```python
plt.figure(figsize=(14, 7))
highest.plot(kind='bar')
plt.title('Number of TV Shows by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()


```

# ii) Compare each countries tv show and movie count and show them in a graph as well

```python
country_groupby = df.groupby(['Country', 'Category']).size().unstack().fillna(0)

# Plotting the results
country_groupby.plot(kind='box', stacked=True, figsize=(14, 7))
plt.title('TV Shows and Movies Count by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```
